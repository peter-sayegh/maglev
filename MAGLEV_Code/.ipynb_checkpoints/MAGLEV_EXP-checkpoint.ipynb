```python
import numpy as np
from scipy.optimize import minimize, root
from scipy.odr import polynomial
from scipy import stats
import matplotlib.pyplot as plt

# Constants
mu_0 = 4 * np.pi * 1e-7  # H/m
g = 9.81  # m/s^2

# Experimental data (in SI units)
m_exp = np.array([0, 5, 15, 20, 26, 31, 36, 43, 51, 63, 71, 85, 90, 103, 117, 124, 135, 166, 180, 192, 217, 259, 289, 307, 322]) * 1e-3  # kg
z_exp = np.array([5, 4.8, 4.3, 4, 3.7, 3.5, 3.3, 2.9, 2.8, 2.8, 2.7, 2.1, 2, 1.9, 1.7, 1.5, 1.5, 1.4, 1.4, 1.3, 1.2, 1, 0.8, 0.5, 0]) * 1e-3  # m


# Fixed geometry parameters
a = 0.015  # m
b = 0.12   # m
m0 = 62e-3  # kg
delta=0.8e-3 #m

def A(z):
    return np.sqrt(a**2 + z**2)

def B(z):
    return np.sqrt(b**2 + z**2)

def Fbar(z):
    """Original complicated F̄(z) function"""
    return (1/(a**2+b**2))*(z * (a * np.arcsinh(a / z) + b * np.arcsinh(b / z) - a * np.arcsinh(a / B(z)) - b * np.arcsinh(b / A(z))) +
            a * b * (np.arctan(z / a) - np.arctan(z / b)) - 
            2 * (a**2) * np.arctan(a * (np.sqrt(a**2 + b**2 + z**2) - np.sqrt(a**2 + b**2)) / (z * (b + np.sqrt(a**2 + b**2)))) -
            z * (A(z) + B(z) - z - np.sqrt(a**2 + b**2 + z**2)) - 
            2 * (b**2) * np.arctan(b * (np.sqrt(a**2 + b**2 + z**2) - np.sqrt(a**2 + b**2)) / (z * (a + np.sqrt(a**2 + b**2)))))

# Exclude the data point where z = 0 to avoid division by zero
mask = z_exp != 0
z_exp_nonzero = z_exp[mask]+delta#
m_exp_nonzero = m_exp[mask]

# Compute Fbar(z_i) for each z_i
Fbar_values = np.array([Fbar(z) for z in z_exp_nonzero])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(Fbar_values, m_exp_nonzero)

# Extract p and q
p = slope
q = intercept

print(f"Estimated coefficient p: {p}")
print(f"Estimated intercept q: {q}")
print(f"R-squared: {r_value**2}")
print(Fbar_values)
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(Fbar_values, m_exp_nonzero, label='Experimental Data', color='blue')
plt.plot(Fbar_values, p * Fbar_values+q, label='Fitted Line', color='red')
plt.xlabel('F̄(z_i)')
plt.ylabel('m_i (kg)')
plt.title('Linear Regression of m_i vs F̄(z_i)')
plt.legend()
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0.015  # m
b = 0.12   # m
mu_0 = 4 * np.pi * 1e-7
delta = 0.8e-3  # m

# Helper functions
def A(z):
    return np.sqrt(a**2 + z**2)

def B(z):
    return np.sqrt(b**2 + z**2)

def Fbar(z):
    """Force function"""
    term1 = z * (a * np.arcsinh(a / z) + b * np.arcsinh(b / z) 
                - a * np.arcsinh(a / B(z)) - b * np.arcsinh(b / A(z)))
    term2 = a * b * (np.arctan(z / a) - np.arctan(z / b))
    term3 = -2 * (a**2) * np.arctan(a * (np.sqrt(a**2 + b**2 + z**2) - np.sqrt(a**2 + b**2)) 
                                     / (z * (b + np.sqrt(a**2 + b**2))))
    term4 = -2 * (b**2) * np.arctan(b * (np.sqrt(a**2 + b**2 + z**2) - np.sqrt(a**2 + b**2)) 
                                     / (z * (a + np.sqrt(a**2 + b**2))))
    return (1 / (a**2 + b**2)) * (term1 + term2 + term3 + term4)

# Generate z values
z_values = np.linspace(0.001, 0.006, 200)
F_values = np.array([Fbar(z) for z in z_values])

# Calculate derivative numerically
dF_dz = np.gradient(F_values, z_values)

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot F(z)
ax1.plot(z_values * 1000, F_values, 'b-', linewidth=2, label='F̃(z)')
ax1.set_xlabel('z (mm)')
ax1.set_ylabel('F̃(z)')
ax1.set_title('Normalized Force F̃(z) vs Height')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Plot dF/dz
ax2.plot(z_values * 1000, dF_dz, 'r-', linewidth=2, label='dF̃/dz')
ax2.axhline(y=0, color='k', linestyle='--', alpha=0.3)
ax2.set_xlabel('z (mm)')
ax2.set_ylabel('dF̃/dz')
ax2.set_title('Force Gradient dF̃/dz vs Height (Stability Criterion)')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()

# Check stability condition
print(f"Stability analysis:")
print(f"At z=2mm: dF/dz = {dF_dz[np.argmin(np.abs(z_values - 0.002))]:.3e}")
print(f"At z=4mm: dF/dz = {dF_dz[np.argmin(np.abs(z_values - 0.004))]:e}")
```

```python
# Stability check for real train
a_real = 0.06  # m
b_real = 153   # m
zc = 0.01      # m (critical height - 1cm per your paper)

# Redefine functions for real train geometry
def A_real(z):
    return np.sqrt(a_real**2 + z**2)

def B_real(z):
    return np.sqrt(b_real**2 + z**2)

def Fbar_real(z):
    """Force function for real train"""
    term1 = z * (a_real * np.arcsinh(a_real / z) + b_real * np.arcsinh(b_real / z) 
                - a_real * np.arcsinh(a_real / B_real(z)) - b_real * np.arcsinh(b_real / A_real(z)))
    term2 = a_real * b_real * (np.arctan(z / a_real) - np.arctan(z / b_real))
    term3 = -2 * (a_real**2) * np.arctan(a_real * (np.sqrt(a_real**2 + b_real**2 + z**2) - np.sqrt(a_real**2 + b_real**2)) 
                                     / (z * (b_real + np.sqrt(a_real**2 + b_real**2))))
    term4 = -2 * (b_real**2) * np.arctan(b_real * (np.sqrt(a_real**2 + b_real**2 + z**2) - np.sqrt(a_real**2 + b_real**2)) 
                                     / (z * (a_real + np.sqrt(a_real**2 + b_real**2))))
    return (1 / (a_real**2 + b_real**2)) * (term1 + term2 + term3 + term4)

# Calculate derivative at critical height
h = 1e-6  # small step for numerical derivative
dF_dz_critical = (Fbar_real(zc + h) - Fbar_real(zc - h)) / (2 * h)

print(f"\n=== STABILITY CHECK FOR REAL EMS TRAIN ===")
print(f"Critical height zc = {zc*1000:.1f} mm")
print(f"Force gradient at zc: dF/dz = {dF_dz_critical:.6e}")
print(f"\nStability condition (Eq. 3): dF/dz > 0")
print(f"Result: dF/dz = {dF_dz_critical:.6e}")

if dF_dz_critical > 0:
    print("✓ Stability condition SATISFIED (positive gradient)")
    print("  → System is stable for vertical perturbations")
else:
    print("✗ Stability condition NOT satisfied (negative gradient)")
    print("  → System is unstable - requires active control")
    
# Plot F(z) around critical height for real train
z_range = np.linspace(0.005, 0.015, 100)  # 5mm to 15mm
F_range = [Fbar_real(z) for z in z_range]
dF_range = np.gradient(F_range, z_range)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: F(z)
ax1.plot(z_range * 1000, F_range, 'b-', linewidth=2)
ax1.axvline(x=zc*1000, color='r', linestyle='--', label=f'Critical height ({zc*1000:.0f} mm)')
ax1.set_xlabel('Height z (mm)')
ax1.set_ylabel('F̃(z)')
ax1.set_title('Force vs Height for Real EMS Train')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Plot 2: dF/dz
ax2.plot(z_range * 1000, dF_range, 'r-', linewidth=2)
ax2.axhline(y=0, color='k', linestyle='--', alpha=0.3, label='Stability threshold')
ax2.axvline(x=zc*1000, color='r', linestyle='--', label=f'Critical height ({zc*1000:.0f} mm)')
ax2.set_xlabel('Height z (mm)')
ax2.set_ylabel('dF̃/dz')
ax2.set_title('Force Gradient (Stability Criterion)')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```
